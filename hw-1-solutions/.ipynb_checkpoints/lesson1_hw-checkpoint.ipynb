3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
    1. окружности,
    2. эллипса,
    3. гиперболы.

1. окружности

```python
R = 5
x1 = np.linspace(-R,R,1000)
y1 = [math.sqrt(R**2 - x_**2) for x_ in x1]
y2 = [-math.sqrt(R**2 - x_**2) for x_ in x1]
plt.axis('equal')
plt.plot(x1, y1)
plt.plot(x1, y2)

```

2. эллипса

```python
a = 4
b = 1
x2 = np.linspace(-a**0.5,a**0.5,1000)
y3 = []
y4 = []
y3 = [ b * (1 - x_**2 / a)**0.5 for x_ in x2]
y4 = [-b * (1 - x_**2 / a)**0.5 for x_ in x2]
plt.axis('equal')
plt.plot(x2, y3)
plt.plot(x2, y4)

```

3. гиперболы

```python
xmin = -20
xmax = 20
dx = 0.1
xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)
ylist = 1 / xlist
plt.plot(xlist, ylist);
```
